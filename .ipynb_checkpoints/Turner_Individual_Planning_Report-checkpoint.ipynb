## Loading in the Data

Here, we'll load in the two files containing our data, and assign them to objects in R. We'll also change the types of some of the columns to be more useful to us before combining the two dataframes into one called combined_data. Once the dataframes have been combined, we're adding a new column that says whether or not the user has actually played, based on if there is a value for their start time. All these things should give us a single tidy dataset that's a good starting point for any analysis we may wish to conduct.

```r
download.file("https://raw.githubusercontent.com/nt8669/DSCI-100-007-24-Project/refs/heads/main/data/players.csv", "data/players.csv")
download.file("https://raw.githubusercontent.com/nt8669/DSCI-100-007-24-Project/refs/heads/main/data/sessions.csv", "data/sessions.csv")

players_data <- read_csv("data/players.csv")
sessions_data <- read_csv("data/sessions.csv")

sessions_data <- sessions_data |>
    mutate(start_time = as.POSIXlt(start_time, format = "%d/%m/%Y %H:%M")) |>
    mutate(end_time = as.POSIXlt(end_time, format = "%d/%m/%Y %H:%M"))
players_data <- players_data |>
    mutate(experience = as.factor(experience)) |>
    mutate(gender = as.factor(gender))

combined_data <- full_join(players_data, sessions_data, by = "hashedEmail") |>
    mutate(has_played = !is.na(start_time))
head(combined_data)
```

## Describing the Data

The combined dataset contains many variables, some from the players data and some from the sessions data. Below, is an explanation for each variable from both datasets.
#### Variables from players.csv
- experience: the players' self-reported experience levels with minecraft ranging from "Beginner" being the least experienced to "Pro" being the most experienced
- subscribe: whether or not the player chose to subscribe to a game related newsletter. Can be TRUE or FALSE
- played_hours: numerical variable, the number of hours the player has spent playing on the server
- name: Not the real names of the participants, rather this variable contains the screen name the players chose from a random list during registration. Variable is a string of characters, not numeric
- gender: gender of the participants, who can select between Male, Female, Two-spirit, Agender, Non-binary, and Prefer not to say
#### Variables from sessions.csv
- start_time: a string of characters giving the date and time of the start of a session
- end_time: a string of characters giving the date and time of the end of a session
- original_start_time: a numerical value that coresponds to the start time of a session, not very useful when we already have start time in a nicer format 
- original_end_time: a numerical value that coresponds to the end time of a session, not very useful when we already have the end time in a nicer format
#### Variables from both datasets
- hashedEmail: a string of characters coresponding to a player's email adress used in registration. This is how we identify individual players, and how we combine the two datasets since it's the only shared variable
#### New variables
- has_played: whether or not the user associated with the hashedEmail variable has ever logged into the server (whether or not they have logged a start time). Can be TRUE or FALSE

## Questions

This project will, broadly, investigate which player characteristics and behaviours are predictive of a given player signing up for the game-related newsletter. Specifically, this project will ask whether a player's total time played, age, and average time of day of day of their play sessions can predict whether or not they will sign up for a newsletter. The first two predictors are relatively simple and require pretty much no wrangling, but the third will be much trickier. First, we will have to find the midpoint of each session using the start and end time. Next, using the combined dataset created near the beginning of this document, we will use the summarize function, grouped by hashEmail to find the mean of these midpoints for each player. This variable, our third predictor, will represent the average time of day at which a given player plays the game.

## Exploratory Analysis and Visualization

```r
players_means <- players_data |>
    select(subscribe, played_hours, Age) |>
    map_df(mean, na.rm = TRUE)
players_means
```

Above are the mean values for each quantitative variable in players.csv. Age and played hours are self explanitory, but the mean in the "subscribe" variable represents the portion of players who subscribed to the newsletter, which is ~73.5%

#### Plots

```r
age_hist <- players_data |>
    ggplot(aes(x = Age)) +
    geom_histogram(color = "dark blue", fill = "light blue", binwidth = 1) +
    labs(x = "User Age (years)", y = "Number of Users", title = "User Age Histogram")
age_hist
```

The most important insight from the plot above is that the majority of players on the server are age 17-24. The reason 17 is such an outlier is that it is the default age set by the website when creating an account, so any users who do not set their age will be listed as 17. 

```r
options(repr.plot.width = 10)

age_v_playtime_low <- players_data |> 
    filter(played_hours < 4) |>
    ggplot(aes(x = Age, y = played_hours)) +
    geom_point(aes(shape = subscribe, color = gender)) +
    labs(y = "Total Playtime (hours)", x = "User Age (years)", title = "Age-Playtime Scatterplot (low playtime)",
         color = "User is Subscribed to Newsletter", shape = "User's Gender")

age_v_playtime_high <- players_data |> 
    filter(played_hours >= 4) |>
    ggplot(aes(x = Age, y = played_hours)) +
    geom_point(aes(shape = subscribe, color = gender)) +
    labs(y = "Total Playtime (hours)", x = "User Age (years)", title = "Age-Playtime Scatterplot (high playtime)",
         color = "User is Subscribed to Newsletter", shape = "User's Gender")

age_v_playtime_low
age_v_playtime_high
```

The two scatterplots above show no clear relationship between age and playtime, other than the fact that people under 30 tend to spend more time playing than people over 30. Notably though, these plots reveal that those with high playtime (> 4 hours) are far more likely to be subscribed to the newsletter than those with low playtime. 

```r
players_bar <- players_data |>
    ggplot(aes(x = gender, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(x = "User's Gender", y = "Portion of Users", 
         fill = "User is Subscribed to Newsletter",
         title = "Gender-Subscription Bar Plot")
players_bar
```

The plot above shows the portions of users of each gender who are subscribed to the newsletter. It shows that users who self-identified as agender or other subscribe 100% of the time, but these sample sizes are so small that those conclusions mean very little. It also shows that people who prefer not to specify their gender are least likely to subscribe. Notably, this plot also shows that women are slightly more likely to subscribe to the newsletter than men, and this conclusion is more meaningful thanks to the relatively large sample sizes of male and female users.

## Methods and Plan

The most appropriate method for investigating whether or not time of day of sessions, total time played, and age are predictive of newsletter subscription would be knn classification. Our question is pretty clearly a classification question, so it makes sense to use the classificaiton method learned in class. The knn model will have to scale and center all of the variables, but doing so requires the assumption that all of the predictor values are equally important in predicting subscription status, which likely isn't true. In order to train the knn model, we will have to split the data (after it has been wrangled) into training and testing, but the exact ratio of that split is something that can be decided later. As for the wrangling itself, we'll need to at least do what we mentioned in the question section, and it may be helpful to select only the columns in the combined dataset that are relevant to our question. Lastly, we will use cross-validation in the training of the model to determine the ideal number of neighbours. Of course the usefulness of the model will likely be limited by the relatively small dataset (less than 200 participants), but there's nothing we can do about that.
